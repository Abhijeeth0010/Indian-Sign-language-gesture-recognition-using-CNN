```python
from PIL import Image
import os
import matplotlib.pyplot as plt

def resize_images(dataset_folder, output_folder, target_size=(1050, 1050)):
    # Create the output folder if it doesn't exist
    os.makedirs(output_folder, exist_ok=True)

    # Lists to store original and resized images
    original_images = []
    resized_images = []

    # Loop through each image in the dataset folder
    for root, dirs, files in os.walk(dataset_folder):
        for filename in files:
            if filename.endswith(('.jpg', '.png', '.jpeg')):
                image_path = os.path.join(root, filename)

                # Open the image using Pillow
                original_img = Image.open(image_path)

                # Resize the image
                resized_img = original_img.resize(target_size)

                # Append the original and resized images to the lists
                original_images.append(original_img)
                resized_images.append(resized_img)

                # Specify the output path for the resized image
                output_path = os.path.join(output_folder, filename)

                # Save the resized image
                resized_img.save(output_path)

    # Plot a graph comparing original and resized images
    num_images = len(original_images)
    max_images_per_figure = 10  # Maximum number of images per figure to avoid large sizes

    for start in range(0, num_images, max_images_per_figure):
        end = min(start + max_images_per_figure, num_images)
        fig, axes = plt.subplots(end - start, 2, figsize=(8, 4 * (end - start)))

        if end - start == 1:
            axes = [axes]

        for i in range(start, end):
            idx = i - start
            axes[idx][0].imshow(original_images[i])
            axes[idx][0].set_title('Original Image')
            axes[idx][0].axis('off')

            axes[idx][1].imshow(resized_images[i])
            axes[idx][1].set_title('Resized Image')
            axes[idx][1].axis('off')

        plt.tight_layout()
        plt.show()

# Specify the path to your dataset folder
dataset_folder = r'C:\Users\Abhijeeth\Desktop\ML\data\Z'

# Specify the path to the output folder for resized images
output_folder = r'C:\Users\Abhijeeth\Desktop\ML\data\Z'

# Resize the images and store them in the output folder, then plot the graph
resize_images(dataset_folder, output_folder)
```

```python
import cv2
import os

# Specify the path to the directory containing images
image_dir = r'C:\Users\Abhijeeth\Desktop\ML\data\G'

# Specify the target format ('RGB' or 'BGR')
target_format = 'RGB'

# Create a new directory to save converted images
output_dir = r'C:\Users\Abhijeeth\Desktop\ML\data\D\RGB'
os.makedirs(output_dir, exist_ok=True)
# Supported image extensions
supported_extensions = ('.jpg', '.jpeg', '.png', '.bmp', '.tiff')

# Iterate through each image in the directory
for image_name in os.listdir(image_dir):
    image_path = os.path.join(image_dir, image_name)
    
    # Check if the file is an image
    if not image_name.lower().endswith(supported_extensions):
        print(f"Skipping non-image file: {image_name}")
        continue

    # Load the image
    original_image = cv2.imread(image_path)

    # Check if the image is loaded successfully
    if original_image is None:
        print(f"Could not read the image at path: {image_path}")
        continue

    # Convert to RGB or BGR
    try:
        if target_format == 'RGB':
            converted_image = cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB)
        elif target_format == 'BGR':
            converted_image = cv2.cvtColor(original_image, cv2.COLOR_RGB2BGR)
        else:
            raise ValueError("Invalid target format. Use 'RGB' or 'BGR'.")
    except Exception as e:
        print(f"Error converting image {image_name}: {e}")
        continue

    # Save the converted image
    output_path = os.path.join(output_dir, image_name)
    try:
        cv2.imwrite(output_path, converted_image)
    except Exception as e:
        print(f"Error saving image {image_name}: {e}")

print(f"All images converted to {target_format} and saved to {output_dir}")
```

```python
import os
import cv2
import matplotlib.pyplot as plt

def threshold_image(input_image_path, output_folder):
    # Read the image
    img = cv2.imread(input_image_path)

    # Convert the image to grayscale
    gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # Apply thresholding
    _, threshold_img = cv2.threshold(gray_img, 90, 255, cv2.THRESH_BINARY_INV)

    # Save the thresholded image to the output folder
    output_path = os.path.join(output_folder, os.path.basename(input_image_path))
    cv2.imwrite(output_path, threshold_img)

    # Optional: Display the thresholded image using matplotlib
    plt.figure(figsize=(6, 6))
    plt.imshow(threshold_img, cmap="gray")
    plt.title(os.path.basename(input_image_path))
    plt.show()

def apply_threshold_to_folder(input_folder, output_folder):
    # Create the output folder if it doesn't exist
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    # Process each image in the input folder
    for filename in os.listdir(input_folder):
        if filename.endswith(('.jpg', '.jpeg', '.png')):
            input_image_path = os.path.join(input_folder, filename)
            threshold_image(input_image_path, output_folder)

# Example usage:
input_folder_path = r'C:\Users\Abhijeeth\Desktop\ML\data\P'
output_folder_path = r'C:\Users\Abhijeeth\Desktop\ML\data\P\Treshold'
apply_threshold_to_folder(input_folder_path, output_folder_path)
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os

def create_skin_mask(image_path, output_folder):
    # Load the image
    original_image = cv2.imread(image_path)

    # Check if the image is loaded successfully
    if original_image is None:
        raise FileNotFoundError(f"Could not read the image at path: {image_path}")

    # Convert the image to the HSV color space
    hsv_image = cv2.cvtColor(original_image, cv2.COLOR_BGR2HSV)

    # Define a range of skin color in HSV
    lower_skin = np.array([0, 20, 70], dtype=np.uint8)
    upper_skin = np.array([20, 255, 255], dtype=np.uint8)

    # Create a mask for the skin color range
    skin_mask = cv2.inRange(hsv_image, lower_skin, upper_skin)

    # Apply the mask to the original image
    skin_detected_image = cv2.bitwise_and(original_image, original_image, mask=skin_mask)

    # Display the images
    plt.figure(figsize=(12, 6))

    plt.subplot(1, 3, 1)
    plt.imshow(cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB))
    plt.title('Original Image')

    plt.subplot(1, 3, 2)
    plt.imshow(skin_mask, cmap='gray')
    plt.title('Skin Mask')

    plt.subplot(1, 3, 3)
    plt.imshow(cv2.cvtColor(skin_detected_image, cv2.COLOR_BGR2RGB))
    plt.title('Skin Detected Image')

    plt.show()

    # Save the skin mask image in a new folder
    os.makedirs(output_folder, exist_ok=True)
    output_path = os.path.join(output_folder, f'skin_mask_{os.path.basename(image_path)}')
    cv2.imwrite(output_path, skin_mask)

    print(f"Skin mask image saved to: {output_path}")

# Specify the path to your image
image_path = r'C:\Users\Abhijeeth\Desktop\ML\data\s\0.jpg'

# Specify the output folder for saving the skin mask image
output_folder = r'C:\Users\Abhijeeth\Desktop\ML\s'
# Create the skin mask, display the results, and save the skin mask image
create_skin_mask(image_path, output_folder)
```

```python
import cv2
import os

# Path to the dataset folder
dataset_folder = r'C:\Users\Abhijeeth\Desktop\ML\data\Z'

# Create a new folder to save rotated images
output_folder = r'C:\Users\Abhijeeth\Desktop\ML\data\Z\Rotated'
os.makedirs(output_folder, exist_ok=True)

# Iterate through all files in the dataset folder
for filename in os.listdir(dataset_folder):
    # Make sure the file is an image (you can add more image file extensions if needed)
    if filename.lower().endswith(('.png', '.jpg', '.jpeg')):
        # Construct the full path to the image
        image_path = os.path.join(dataset_folder, filename)

        # Read the original image
        original_img = cv2.imread(image_path)

        # Check if the image is loaded successfully
        if original_img is not None and len(original_img.shape) == 3:
            # Rotate the image by 90 degrees anticlockwise
            rotated_img = cv2.rotate(original_img, cv2.ROTATE_90_COUNTERCLOCKWISE)

            # Display the rotated image (optional)
            # cv2.imshow('Rotated Image', rotated_img)
            # cv2.waitKey(0)
            # cv2.destroyAllWindows()

            # Save the rotated image to the output folder
            output_filename = f"rotated_{filename.split('.')[0]}_90_anticlockwise.png"
            output_path = os.path.join(output_folder, output_filename)
            cv2.imwrite(output_path, rotated_img)

print("Rotation (90 degrees anticlockwise) completed.")
```

```python
from sklearn.model_selection import train_test_split
import os
import shutil

def split_dataset(original_dataset_folder, output_folder, test_size=0.2, random_state=42):
    # Create train and test folders
    train_folder = os.path.join(output_folder, 'train')
    test_folder = os.path.join(output_folder, 'test')
    os.makedirs(train_folder, exist_ok=True)
    os.makedirs(test_folder, exist_ok=True)

    # Iterate through subfolders in the original dataset folder
    for subfolder_name in os.listdir(original_dataset_folder):
        subfolder_path = os.path.join(original_dataset_folder, subfolder_name)
        
        # Check if it's a directory
        if os.path.isdir(subfolder_path):
            # List all image files in the subfolder
            image_files = [f for f in os.listdir(subfolder_path) if f.endswith(".jpg")]

            # Split the dataset into train and test sets for the current subfolder
            train_images, test_images = train_test_split(image_files, test_size=test_size, random_state=random_state)

            # Move images to the respective train and test folders
            for image_file in train_images:
                src_path = os.path.join(subfolder_path, image_file)
                dest_path = os.path.join(train_folder, subfolder_name, image_file)
                os.makedirs(os.path.join(train_folder, subfolder_name), exist_ok=True)
                shutil.copy(src_path, dest_path)

            for image_file in test_images:
                src_path = os.path.join(subfolder_path, image_file)
                dest_path = os.path.join(test_folder, subfolder_name, image_file)
                os.makedirs(os.path.join(test_folder, subfolder_name), exist_ok=True)
                shutil.copy(src_path, dest_path)

    print(f"Dataset split into train and test sets. Train set: {len(train_images)} images, Test set: {len(test_images)} images.")

# Specify the path to your original dataset folder
original_dataset_folder = r'C:\Users\Abhijeeth\Desktop\ML\data'

# Specify the path to the folder where you want to create the train and test sets
output_folder = r'C:\Users\Abhijeeth\Desktop\ML\data_split'

# Split the dataset
split_dataset(original_dataset_folder, output_folder)
```
